```python
#Data Visualization 2

from pandas import *
from ggplot import *

def plot_weather_data(turnstile_weather):
    '''

    Here are some suggestions for things to investigate and illustrate:
     * Ridership by time of day or day of week
     * How ridership varies based on Subway station
     * Which stations have more exits or entries at different times of day
     
    You can check out:
    https://www.dropbox.com/s/meyki2wl9xfa7yk/turnstile_data_master_with_weather.csv
    '''
    turnstile_weather = pandas.read_csv(turnstile_weather)
    turnstile_weather = turnstile_weather[['UNIT',"ENTRIESn_hourly"]].groupby('UNIT', as_index=False).mean()
    turnstile_weather=turnstile_weather.reset_index()
    
    sortedEntries = turnstile_weather.sort(["ENTRIESn_hourly", "UNIT"], ascending=False)
    sortedEntries=sortedEntries.head(14)
    sortedEntries=sortedEntries.reset_index()
    
    plot = ggplot(sortedEntries, aes(x='UNIT', y='ENTRIESn_hourly')) + \
        geom_bar(aes(weight='ENTRIESn_hourly'),fill='lightblue',stat = 'identity') + \
        ggtitle('How ridership varies by subway station') + xlab('Station') + ylab('Entries')
    
    print plot

if __name__== '__main__':
    plot_weather_data('data/turnstile_data_master_with_weather.csv')
```

```python
from pandas import *
from ggplot import *

def plot_weather_data(turnstile_weather):
    '''

    Here are some suggestions for things to investigate and illustrate:
     * Ridership by time of day or day of week
     * How ridership varies based on Subway station
     * Which stations have more exits or entries at different times of day
     
    You can check out:
    https://www.dropbox.com/s/meyki2wl9xfa7yk/turnstile_data_master_with_weather.csv
    '''
    turnstile_weather = pandas.read_csv(turnstile_weather)
    turnstile_weather = turnstile_weather.reset_index()
    with_rain = turnstile_weather['ENTRIESn_hourly'][turnstile_weather['rain']==1]
    without_rain = turnstile_weather['ENTRIESn_hourly'][turnstile_weather['rain']==0]
    #print len(with_rain)
    print with_rain
    
    
    turnstile_weather.DATEn = pandas.to_datetime(turnstile_weather.DATEn)
    f = lambda x:int(x[1:])
    turnstile_weather['station'] = turnstile_weather['UNIT'].map(f)
    
    station_part1 = turnstile_weather[turnstile_weather['station']>=10]
    station_part2 = station_part1[station_part1['station']<20]
    
    filteredCols = station_part2[['UNIT', 'station', 'DATEn', 'ENTRIESn_hourly']]
    #print filteredCols
    grouped = filteredCols.groupby(['UNIT','station', 'DATEn'], as_index=False).sum()
    #print grouped
    #print turnstile_weather.station.head(50)
    #testplot = ggplot(aes(x='ENTRIESn_hourly'), data = turnstile_weather) + geom_histogram()
    
    plot1 = ggplot(grouped, aes(x='DATEn', y='ENTRIESn_hourly', color='UNIT')) + \
            geom_line() + \
            ggtitle('How ridership varies by subway station')  + \
            xlab('Date') + ylab('Entries') + \
            theme(axis_text_x = element_text(angle = 45, hjust = 1))
            
        
    print plot1
    
if __name__== '__main__':
    plot_weather_data('data/turnstile_data_master_with_weather.csv')
```
